```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### APP RATINGS GRAPH

```python
plt.hist(df["Rating"].dropna(),bins=20)
plt.title("Distribution of app Ratings")
plt.xlabel("Ratings")
plt.ylabel("count")
plt.show()
```

This histogram visualizes how app rating are spread accross playstore dataset.most apps cluster between 4.0 and 4.5 indicating generally higher user satisfaction.lower ratings are rare while peak counts arounde higher ratings suggest that user tend to rate apps positivily

## REVIEWS VS RATING

```python
plt.scatter(df['Rating'],df['Reviews'])
plt.xlabel('Rating')
plt.ylabel('Reviews')
plt.title('Reviews vs Rating')
plt.show()
```

This scatterplot shows the relationship between the number of reviews and the app rating in the playstore data set there is no strong corrilation between the number of reviews and app ratings,
although highly reviewed apps to tends to maintain good ratings


## INSTALLS VS RATING

```python
plt.scatter(df['Rating'],df['Installs'])
plt.xlabel('Rating')
plt.ylabel('Installs')
plt.title('Installs vs Rating')
plt.show()
```

## FREE VS PAID APPS

```python
freq=df['Type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('graph of column of Type')
plt.ylabel('')
plt.show()
```

this "Free" category take up the vast majority of the chart,accounting for 93.11% of the total apps.the"paid" category apps have very small 6.89 share



## CORRELATION HEATMAP

```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.title("correlation heatmap")
plt.show()
```

in this data ,the only thing that really moves together is the number of installs and the number of reviews.other things,like how much an app costs or its size dont seemto affect the user thing rating much at all

## RATING COMPARISON BY APP TYPE

```python
sns.boxplot(x='Type',y='Rating',data=df)
plt.title('Rating comparison by app type')
```

find most apps for free,paid apps tends to have small edge in overall user satisfaction

## TOP 10 APP CATEGORY

```python
df['Category'].value_counts().head(10).plot(kind='bar')
plt.title('app categories')
plt.xlabel('categories')
plt.ylabel('number of apps')
plt.title('top 10 app categories')
plt.show()
```

this bar chart,titled "Top 10 categories",provides a visual representation of the most populated categories in the data set based on the total number of apps 
family having the highest count(nearly 1750)and the other categories ranging down to the lowest(MEDICAL AND PHOTOGRAPHY,250 each)


```python
reviews_install=df.groupby('Category')[['Reviews','Installs']].mean()
```

```python
reviews_install
```

## REVIEWS VS INSTALLS

```python
plt.scatter(reviews_install["Reviews"],reviews_install["Installs"])
plt.xlabel("Reviews")
plt.ylabel("Installs")
plt.title("Reviews vs installs")
plt.show()
```

this graph confirms a strong relationship where higher installs lead to higher volume of reviews

## REVIEWS VS INSTALLS (LINE GRAPH)

```python
sns.lineplot(data=reviews_install,x="Reviews",y="Installs")
```

```python
reviews_rating=df.groupby('Category')[['Reviews','Rating']].mean()
```

```python
reviews_rating.sort_values(by ="Reviews",ascending = False)
```

```python
reviews_rating.head(10)
```

## TOP 10 CATEGORIES AVERAGE RATING

```python
reviews_rating.head(10)["Rating"].plot(kind="bar",figsize=(10,5))
plt.xlabel("Category")
plt.ylabel("Avarage rating")
plt.title("top 10 categories avarage rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
